#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código

calificaciones = [8.5, 9.2, 7.8, 8.9, 10]
suma = sum(calificaciones)
longitud = len(calificaciones)
promedio = suma / longitud


print(f"El promedio de la lista es: {promedio}")
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
calificacion_mas_alta = max(calificaciones)
calificacion_mas_baja = min(calificaciones)

print(f"la calificacion mas alta es: {calificacion_mas_alta}")
print(f"la calificacion mas baja es: {calificacion_mas_baja}")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO


ds_precio_minimo_filtrado = ds[ds["price"] != 0]
indice_minimo = ds_precio_minimo_filtrado["price"].idxmin()

precio_minimo = ds_precio_minimo_filtrado["price"].min()

direccion_minimo = ds.loc[indice_minimo, "address"]

print(f"la casa con direccion en: {direccion_minimo} es la mas barata y su precio es: {precio_minimo}")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# TODO

columna_direccion = "address"
superficie = "surface"

#para encontrar casa mas grande.

superficie_grande = ds[superficie].max()
indice_max = ds[superficie].idxmax()

direccion_max = ds.loc[indice_max, columna_direccion]


#para encontrar casa mas pequeña.

superficie_pequeña = ds[superficie].min()
indice_min = ds[superficie].idxmin()

direccion_min= ds.loc[indice_min, columna_direccion]



print(f"La casa mas grande esta ubicada en: {direccion_max}, y su superficie es de: {superficie_grande}")
print(f"La casa mas pequeña esta ubicada en: {direccion_min}, y su superficie es de: {superficie_pequeña}")
```
